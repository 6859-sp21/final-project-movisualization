```python
import pandas as pd
```

```python
imdb_movies_path = 'imdb_data/IMDb movies.csv'
imdb_ratings_path = 'imdb_data/IMDb ratings.csv'
movie_path = 'data_cleaned_merged/movie_title_metadata.csv'

imdb_df = pd.read_csv(imdb_movies_path)
ratings_df = pd.read_csv(imdb_ratings_path)
movies_df = pd.read_csv(movie_path)
ratings_df
```

```python
imdb_df['title'] = imdb_df['title'].apply(lambda x: x.lower().title())
```

```python
merged = ratings_df.merge(imdb_df, on='imdb_title_id')
```

```python
movies_df.merge(merged, on='title',how='left')
```
