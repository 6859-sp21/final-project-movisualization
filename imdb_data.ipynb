```python
import pandas as pd
import string
```

```python
imdb_movies_path = 'imdb_data/IMDb movies.csv'
imdb_ratings_path = 'imdb_data/IMDb ratings.csv'
movie_path = 'data_cleaned_merged/movie_title_metadata.csv'

imdb_df = pd.read_csv(imdb_movies_path)
ratings_df = pd.read_csv(imdb_ratings_path)
movies_df = pd.read_csv(movie_path)

print(imdb_df.columns)
```

```python
merged_imdb = ratings_df.merge(imdb_df, on='imdb_title_id')
merged_imdb['year'] = merged_imdb['year'].apply(lambda x: str(x)[0:4])
merged_imdb['original_title'] = merged_imdb['original_title'].apply(lambda x: x.translate(None, string.punctuation).lower().title())
merged_imdb = merged_imdb.rename(columns={'title':'title_spanish'})
merged_imdb = merged_imdb.rename(columns={'original_title':'title'})

movies_df['title'] = movies_df['title'].apply(lambda x: x.translate(None, string.punctuation).lower().title())
movies_df['year'] = movies_df['year'].apply(lambda x: str(x)[0:4])

movies_df.head()
```

```python
all_data = movies_df.merge(merged_imdb, on=['title','year'])
                            
```

```python
movies = pd.read_csv(movie_path)[['movie_id','title']]
all_imdb_final = all_data.drop(columns='title')
final_df = movies.merge(all_imdb_final, on=['movie_id'])
# final_df
```

```python
unmerged = all_data[all_data['imdb_title_id'].isna()][['movie_id', 'title', 'year']]
unmerged['title'].tolist()
```
